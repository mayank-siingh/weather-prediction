We first preprocess data to convert it from pandas dataframe to numpy array so as to operate scikit-learn

```python
X = df[[' _dewptm',' _fog',' _hail',' _heatindexm',' _hum',' _pressurem',' _rain',' _snow',' _tempm',' _thunder',' _tornado',' _vism',' _wdird',' _wdire',' _wgustm',' _windchillm',' _wspdm']]
y = df[[' _conds']] # taking a parameter as target , ofcourse it can be any other attribute as well
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
X = preprocessing.StandardScaler().fit(X).transform(X)
```

Now dividing the dataset into training and testing datasets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=4)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=4)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
predict = model.predict(X_test)
print(confusion_matrix(predict,y_test))
print(classification_report(predict,y_test))
```

Getting the accuracy score of the Decision Tree 

```python
from sklearn import metrics
print("Decision Tree Accuracy: ",metrics.accuracy_score(y_test,predict))
```

Now predicting using Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 200)
rfc.fit(X_train,y_train)
```

```python
rfc_predict = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfc_predict))
print(classification_report(y_test,rfc_predict))
```

```python
print("Random Forest Accuracy: ",metrics.accuracy_score(y_test,rfc_predict))
```
