Importing pandas for converting a downloaded csv file into dataframe

```python
import pandas as pd
df = pd.read_csv('testset.csv')
df
```

dropping unnecessary columns

```python
columns = [' _conds',' _heatindexm',' _precipm',' _snow',' _thunder',' _tornado',' _vism',' _wdire',' _wgustm',' _windchillm',' _wspdm']
df.drop(columns,axis=1,inplace=True)
df
```

dropping rows with NaN values 

```python
df.dropna(inplace=True)
df
```

Dividing the dataset into training and testing sets

```python
from sklearn.model_selection import train_test_split
X = df[[' _dewptm',' _fog',' _hail',' _hum',' _pressurem',' _rain',' _wdird']]
y = df[[' _tempm']]
train_X,test_X,train_y,test_y = train_test_split(X,y,train_size=0.8,random_state=1)
```

We will use decision tree regressor model 

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
```

```python
model.fit(train_X,train_y)
```

```python
from sklearn.metrics import mean_absolute_error
predict_y = model.predict(test_X)
mean_absolute_error(predict_y,test_y)
```
